$$
\huge f(x) = k x (1 - x)
$$


```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
def logistic_equation(k, x):
    return(k * x * (1 - x))

vect_logistic_equation = np.vectorize(logistic_equation)
```

```python
def iterate(k, x, num_steps = 100):
    
    for i in range(1, num_steps):
        x = vect_logistic_equation(k, x)
        
    return(x)

vect_iterate = np.vectorize(iterate)
```

```python
ks = np.arange(2,4,.0001)
x = .5

ys = vect_iterate(ks, x)
    
```

```python
plt.figure(figsize=[10, 8])
plt.scatter(ks, ys, c='black', marker='o', s=.5);
```
